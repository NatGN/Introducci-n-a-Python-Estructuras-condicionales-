Parte 2: Limpieza de Datos
1. Muestra la cantidad de valores nulos y eliminar valores faltantes

```python
df.isnull().sum()
```

```python
df_sin_nulos = df.dropna()
```

2. Rellenar valores faltantes: Rellena los valores faltantes con 0.


```python
df_relleno = df.fillna(0)
```

3. Eliminar columnas innecesarias: Elimina una columna que no sea relevante para el análisis.


```python
df = df.drop(columns=["sexo"])
```

Parte 3: Filtrado y Selección de Datos
1. Filtrar con múltiples condiciones: Filtra las filas que cumplan con múltiples condiciones.

```python
df_filtrado = df[
    (df["edad"] >= 20) &
    (df["carrera"] == "Ingeniería de Sistemas") &
    (df["acepta"] == "Si")
]

df_filtrado
```

2. Seleccionar columnas: Selecciona solo ciertas columnas para un nuevo DataFrame.


```python
df_columnas = df[["carrera", "edad", "trabajo"]]
df_columnas
```

Parte 4: Operaciones y Agrupación
1. Calcula la suma, media de una columna

```python
suma_edad = df["edad"].sum()
media_edad = df["edad"].mean()

print("Suma de edades:", suma_edad)
print("Edad promedio:", media_edad)
```

2. Agrupa los datos por una columna y calcula la media de otra columna para cada grupo.


```python
edad_por_carrera = df.groupby("carrera")["edad"].mean()
edad_por_carrera
```

3. Crea una tabla pivote para resumir los datos.


```python
tabla_pivote = pd.pivot_table(
    df,
    values="edad",
    index="carrera",
    aggfunc="mean"
)

tabla_pivote
```

Ejercicio 5: Manipulación de Datos
1. Ordenar datos: Ordena el DataFrame por una columna específica.


```python
df_ordenado = df.sort_values(by="edad", ascending=False)
df_ordenado
```

Ejercicio 6: Guardar Resultados
1. Guardar DataFrame: Guarda el DataFrame modificado en un nuevo archivo CSV.
